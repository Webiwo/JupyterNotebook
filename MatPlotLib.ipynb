```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.show()
```

```python
# 1st method
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# 2nd method

# Create some data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
# 3rd method (recommended)

fig, ax = plt.subplots(figsize=(10,3)) # (width, height)
ax.plot(x, y)
ax.set(
    title="Simple Plot",
    xlabel="x-axis",
    ylabel="y-axis"
)
plt.show()
#plt.savefig("images/simple-plot.png")    
```

### Subplots

```python
# Option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10, 5))
ax1.plot(x, 2*x)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
plt.show()
```

```python
# Option 2
fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10, 5))
ax[0, 0].plot(x, 2*x)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
plt.show()
```

### Heart Disease

```python
hd = pd.read_csv("heart-disease.csv")
hd[:10]
```

```python
# Create a histogram of age
hd["age"].plot(kind="hist", bins=20)
plt.show()
```

## OO method

```python
over_50 = hd[hd["age"] > 50 ]
over_50.head()
```

```python
len(over_50)
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax)
ax.set_xlim([45, 100])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="age",
       ylabel="cholesterol")
ax.legend(*scatter.legend_elements(), title="Target")
ax.axhline(y=over_50["chol"].mean(), color="r", linestyle="--", label="Avg Chol")
plt.show()                     
```

```python
# subplots of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10))
#add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])
ax0.set(title="Heart Disease and Cholesterol Levels",
       xlabel="age",
       ylabel="cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(), color="r", linestyle="--")

#add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="age",
       ylabel="Max Heart Rate")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50["thalach"].mean(), color="r", linestyle="--")

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
plt.show()
```

```python
plt.style.use("seaborn-v0_8-whitegrid")
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)
#add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")
ax0.set(title="Heart Disease and Cholesterol Levels",
       xlabel="age",
       ylabel="cholesterol")
ax0.set_xlim(50,80)
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_50["chol"].mean(), color="r", linestyle="--")

#add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter")
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="age",
       ylabel="Max Heart Rate")
ax1.set_ylim(60,200)
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_50["thalach"].mean(), color="r", linestyle="--")

fig.suptitle("Heart Disease Analysis", fontsize=14, fontweight="bold")
plt.show()
```
